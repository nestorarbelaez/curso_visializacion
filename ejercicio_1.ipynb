posición Y (Variable numércia)

*   salario total del equipo
*   posición del equipo medido como la comparación entre triunfos y perdidas de los partidos de beibol



Posición X (variable categórica)

* separación de ligas



Ancho (variable numérica)

* Indicación del salario comparado los otros equipos

Color (categórica)

Manejo del equipo, es decir posición en la tabla comparado con sus gastos relativo a otros equipos

* Rojo para mal manejo
* Azul para buen manejo.



Orientación (categórica)

Según el manejo del equipo se tendra una orientación u otra de las líneas que unen la posición de la tabla con su gasto en nomina.

* Pendiente positiva, mal manejo.
* Pendiente negativa, buen manejo.

Movimiento (categórico)

Cambio del grafico con el tiempo en diferentes jerarquía.
