# Punto 3

Se puede observar una tendencia decreciente de la edad de los diferentes lideres, es decir, cada vez más los mandatiros suelen ser más jóvenes.

En general, se observa que los lideres republicanos suelen ser más viejos que sus pares democratas a excepeción de Biden que sería el presidente estadounidense más viejo de los últimos 70 años.

En los últimos 70 años, el número de presidentes democratas y republicanos ha estado muy equilibrado. Sin embargo, la extensión de sus mandatos es ligeramente superior en el cado de los republicanos.
